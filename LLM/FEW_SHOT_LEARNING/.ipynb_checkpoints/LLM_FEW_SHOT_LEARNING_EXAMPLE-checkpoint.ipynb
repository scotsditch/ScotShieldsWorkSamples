```python
# openai version

def openai_response(prompt,apikey):
    
    import openai
    
    openai.api_key =apikey
    
    response = openai.Completion.create(
    engine='text-davinci-003',
        prompt=prompt,
        max_tokens=1000
    )
    
    return response.choices[0].text.strip()
```
