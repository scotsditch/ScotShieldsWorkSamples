### Naive Predictor:


```python
# Create naive predictors
# always predict is_bad=1


n_positive = len(data_df[data_df['is_bad']==1.0])
n_records = len(data_df)
n_at_most = n_records - n_positive

# Create naive predictor

# TODO: Calculate accuracy
accuracy = float(n_positive)/n_records

# TODO: Calculate F-score using the formula above for beta = 0.5

TP = n_positive # "True Positive" predictions (correct)
FP = n_at_most # "False Positive" predictions (wrong)
FN = 0 # Never predict a 0.0

precission = accuracy # since naive predictor predicts all 1.0

recall = TP/float(TP+FN)

BETA = 1

fscore = (1 + BETA**2)*precission*recall/(BETA**2*precission + recall)

# Print the results 
print("Naive Predictor: [Accuracy score: {:.4f}, F-score: {:.4f}]".format(accuracy, fscore))
```

### Initial model evaluation:

```python


# TODO: Import two metrics from sklearn - fbeta_score and accuracy_score
from sklearn.metrics import fbeta_score, accuracy_score
#from sklearn.utils import shuffle

def train_predict(learner, sample_size, X_train, y_train, X_test, y_test): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - sample_size: the size of samples (number) to be drawn from training set
       - X_train: features training set
       - y_train: label training set
       - X_test: features testing set
       - y_test: label testing set
    '''
    
    results = {}
    
    
    # TODO: Fit the learner to the training data using slicing with 'sample_size'
    start = time() # Get start time
    learner.fit(X_train[:sample_size],y_train[:sample_size])
    end = time() # Get end time
    
    # TODO: Calculate the training time
    results['train_time'] = end-start
        
    # TODO: Get the predictions on the test set,
    #       then get predictions on the first 300 training samples
    start = time() # Get start time
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train[:300])
    end = time() # Get end time
    
    # TODO: Calculate the total prediction time
    results['pred_time'] = end-start
            
    # TODO: Compute accuracy on the first 300 training samples
    results['acc_train'] = accuracy_score(y_train[:300],predictions_train)
        
    # TODO: Compute accuracy on test set
    results['acc_test'] = accuracy_score(y_test,predictions_test)
    
    # TODO: Compute F-score on the the first 300 training samples
    results['f_train'] = fbeta_score(y_train[:300],predictions_train,beta=0.5)
        
    # TODO: Compute F-score on the test set
    results['f_test'] = fbeta_score(y_test,predictions_test,beta=0.5)
       
    # Success
    print("{} trained on {} samples.".format(learner.__class__.__name__, sample_size))
        
    # Return the results
    return results
```

```python

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier


clf_A = LogisticRegression(random_state=42)
clf_B = GradientBoostingClassifier(random_state=42)


samples_1 = int(round(len(X_train)/100))
samples_10 = int(round(len(X_train)/10))
samples_100 = len(y_train)

# Collect results on the learners
results = {}
for clf in [clf_A, clf_B]:#clf_C]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    for i, samples in enumerate([samples_1, samples_10, samples_100]):
        results[clf_name][i] = \
        train_predict(clf, samples, X_train, y_train, X_test, y_test)

# Run metrics visualization for the three supervised learning models chosen
evaluate(results, accuracy, fscore)

```

### K-fold CV Train/Test data

```python
from sklearn.metrics import (roc_auc_score, confusion_matrix,
                             accuracy_score, roc_curve,
                             precision_recall_curve, f1_score, log_loss)
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.preprocessing import Imputer, RobustScaler, FunctionTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer
```

```python
def pred_Kfold(cf,xtr,ytr,scr,flds):
    
    scores = cross_val_score(cf,
                             xtr, ytr,
                             scoring=make_scorer(scr), cv=flds)
    return scores.mean()
```

### Model evaluation:

```python


# K-folds

#### F1score

print('F1score logistic regression: {}'\
      .format(pred_Kfold(cf= LogisticRegression(random_state=42)\
                         ,xtr=X_train,ytr=y_train,scr=f1_score,flds=5)))

print('F1score GradientBoostingClassifier: {}'\
      .format(pred_Kfold(cf= GradientBoostingClassifier(random_state=42)\
                         ,xtr=X_train,ytr=y_train,scr=f1_score,flds=5)))


#### LogLoss

print('LogLoss logistic regression: {}'\
      .format(pred_Kfold(cf= LogisticRegression(random_state=42)\
                         ,xtr=X_train,ytr=y_train,scr=log_loss,flds=5)))

print('LogLoss GradientBoostingClassifier: {}'\
      .format(pred_Kfold(cf= GradientBoostingClassifier(random_state=42)\
                         ,xtr=X_train,ytr=y_train,scr=log_loss,flds=5)))
```

#### Observation:
- F1score and LogLoss are very close for both models.

### Tuning Champion model

```python
# tuning champion model LogisticRegression

# TODO: Import 'GridSearchCV', 'make_scorer', and any other necessary libraries

# from sklearn.grid_search import GridSearchCV
from sklearn.model_selection import GridSearchCV



# TODO: Initialize the classifier
clf = LogisticRegression(random_state=42)

# TODO: Create the parameters list you wish to tune
parameters = {'penalty': ['l1', 'l2'],
              'C':np.logspace(0, 4, 10)}

# TODO: Make an fbeta_score scoring object
scorer = make_scorer(f1_score)

# TODO: Perform grid search on the classifier using 'scorer' as the scoring method
grid_obj = GridSearchCV(clf,parameters,scoring=scorer)

# TODO: Fit the grid search object to the training data and find the optimal parameters
grid_fit = grid_obj.fit(X_train,y_train)

# Get the estimator
best_clf = grid_fit.best_estimator_

# Make predictions using the unoptimized and model
predictions = (clf.fit(X_train, y_train)).predict(X_test)
best_predictions = best_clf.predict(X_test)

# Report the before-and-afterscores
print("Unoptimized model\n------")
print("LogLoss score on testing data: {:.4f}".format(log_loss(y_test, predictions)))
print("F1-score on testing data: {:.4f}".format(f1_score(y_test, predictions)))
print("\nOptimized Model\n------")
print("Final LogLoss score on the testing data: {:.4f}".format(log_loss(y_test, best_predictions)))
print("Final F1-score on the testing data: {:.4f}".format(f1_score(y_test, best_predictions)))
```
